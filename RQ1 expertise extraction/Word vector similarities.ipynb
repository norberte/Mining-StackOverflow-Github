```
from gensim.models.keyedvectors import KeyedVectors
word_vectors = KeyedVectors.load("/content/drive/My Drive/embeddings/SO_pre-trained_vectors.kv", mmap='r')
```

```
words=['python','java','mysql']
for w in words:
    try:
        print(word_vectors.most_similar(w,topn=30))
    except KeyError as e:
            print(e) 
```

```
print(word_vectors.doesnt_match("java c++ python bash".split()))
```

```
print(word_vectors.most_similar(positive=['python', 'eclipse'], negative=['java']))
```

## N_similarities

```
annotation = ['paths', 'usb', 'css', 'java', 'python', 'serial', 'javascript', 'sql', 'private', 'embedded', 'design', 'website', 'c++', 'google', 'html', 'json', 'bash', 'selectors', 'trademark', 'reader', 'c', 'device', 'mysql', 'dotnet', 'rust', 'database', 'maps', 'rfid', 'hid', 'php', 'hal', 'aspdotnet', 'canvas', 'graphics', 'xml', 'cortex', 'regex']  
model_hypothesis = ['python', 'ruby', 'php', 'scala', 'clojure', 'javascript', 'haskell', 'javascript', 'c', 'rust', 'java', 'perl', 'spring', 'c++', 'rails', 'docker', 'shell', 'r', 'ansible', 'awesome', 'elm', 'react', 'django', 'elixir', 'spark', 'kotlin', 'ocaml', 'php', 'java', 'css', 'library', 'boot', 'lisp', 'plugin', 'purescript', 'github', 'theme', 'io', 'erlang', 'data', 'typescript', 'nt', 'nodejs', 'list', 'bundle', 'node']


sim = word_vectors.n_similarity(annotation, model_hypothesis)
print("{:.4f}".format(sim))
```
