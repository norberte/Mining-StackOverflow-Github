```
GH_labels_1 = GH_annotation["Annotator_1"]
GH_labels_2 = GH_annotation["Annotator_2"]
SO_labels_1 = SO_annotation["Annotator_1"]
SO_labels_2 = SO_annotation["Annotator_2"]
```

```
def process_str(list_of_labels):
  processed_label_str = ""
  for label in list_of_labels:
    processed_label = initial_cleaning(label)
    if processed_label.startswith(" "):
      processed_label = processed_label[1:]
    if processed_label.endswith(" "):
      processed_label = processed_label[0:len(processed_label)-1]
    processed_label_str = processed_label_str + processed_label.replace(" ", "_") + ";"
  return processed_label_str
```

```
def process_strings(list_of_labels):
  processed_label_str = ""
  for label in list_of_labels:
    processed_label = initial_cleaning(label)
    if processed_label.startswith(" "):
      processed_label = processed_label[1:]
    if processed_label.endswith(" "):
      processed_label = processed_label[0:len(processed_label)-1]
    processed_label_str = processed_label_str + processed_label.replace(" ", ";") + ";"
  return processed_label_str
```

```
GH_Processed_Annotations1 = []
GH_Processed_Annotations2 = []
SO_Processed_Annotations1 = []
SO_Processed_Annotations2 = []

from gensim.parsing.preprocessing import remove_stopwords

for i in range(0, len(GH_labels_1)):
  temp = standardize_strings(GH_labels_1[i].lower())
  temp = remove_stopwords(temp)
  list_of_labels = temp.replace('"', "").replace('-', ",").split(",")
  processed_label_str = process_strings(list_of_labels)
  GH_Processed_Annotations1.append(processed_label_str)

GH_annotation["Processed_Annotator_1"] = GH_Processed_Annotations1


for i in range(0, len(GH_labels_2)):
  temp = standardize_strings(GH_labels_2[i].lower())
  temp = remove_stopwords(temp)
  list_of_labels = temp.replace('"', "").replace('-', ",").split(",")
  processed_label_str = process_strings(list_of_labels)
  GH_Processed_Annotations2.append(processed_label_str)

GH_annotation["Processed_Annotator_2"] = GH_Processed_Annotations2



for i in range(0, len(SO_labels_1)):
  temp = standardize_strings(SO_labels_1[i].lower())
  temp = remove_stopwords(temp)
  list_of_labels = temp.replace('"', "").replace('-', ",").split(",")
  processed_label_str = process_strings(list_of_labels)
  SO_Processed_Annotations1.append(processed_label_str)

SO_annotation["Processed_Annotator_1"] = SO_Processed_Annotations1


for i in range(0, len(SO_labels_2)):
  temp = standardize_strings(SO_labels_2[i].lower())
  temp = remove_stopwords(temp)
  list_of_labels = temp.replace('"', "").replace('-', ",").split(",")
  processed_label_str = process_strings(list_of_labels)
  SO_Processed_Annotations2.append(processed_label_str)

SO_annotation["Processed_Annotator_2"] = SO_Processed_Annotations2
```
