```
with open('/content/drive/My Drive/EMT/SO_past/vocab.pkl', 'rb') as f:
    vocab = pickle.load(f)
vocab20 = [i for i in vocab.itervalues()]

with open('/content/drive/My Drive/EMT/SO_past/vocab.pkl', 'rb') as f:
    vocab = pickle.load(f)
vocab50 = [i for i in vocab.itervalues()]

embeddings_k20 = model20.rho.data.cpu().numpy() 
embeddings_k50 = model50.rho.data.cpu().numpy() 

topic_embeddings_k20 = model20.alphas.weight.cpu().detach().numpy()
topic_embeddings_k50 = model50.alphas.weight.cpu().detach().numpy()

vectors_k20 = np.concatenate((embeddings_k20, topic_embeddings_k20))
vectors_k50 = np.concatenate((embeddings_k50, topic_embeddings_k50))
```

```
numberOfTopics = 20
for k in range(0, numberOfTopics):
  vocab20.append('T_' + str(k+1))

numberOfTopics = 50
for k in range(0, numberOfTopics):
  vocab50.append('T_' + str(k+1))
```

```
def nearest_neighbors(word, vectors, vocab):
    index = vocab.index(word)
    query = vectors[index]
    ranks = vectors.dot(query).squeeze()
    denom = query.T.dot(query).squeeze()
    denom = denom * np.sum(vectors**2, 1)
    denom = np.sqrt(denom)
    ranks = ranks / denom
    mostSimilar = []
    [mostSimilar.append(idx) for idx in ranks.argsort()[::-1]]
    nearest_neighbors = mostSimilar[:20]
    nearest_neighbors = [vocab[comp] for comp in nearest_neighbors]
    return nearest_neighbors
```

```
def unique_nearest_neighbors(word, vectors, vocab):
    index = vocab.index(word)
    query = vectors[index]
    ranks = vectors.dot(query).squeeze()
    denom = query.T.dot(query).squeeze()
    denom = denom * np.sum(vectors**2, 1)
    denom = np.sqrt(denom)
    ranks = ranks / denom
    mostSimilar = []
    [mostSimilar.append(idx) for idx in ranks.argsort()[::-1]]
    nearest_neighbors = mostSimilar[:50]
    nearest_neighbors = list(set([vocab[comp] for comp in nearest_neighbors]))
    return nearest_neighbors
```

```
results_k20 = []
results_k50 = []
for i in range(1,21):
  results_k20.append(nearest_neighbors('T_' + str(i), vectors_k20, vocab20))

for i in range(1,51):
  results_k50.append(nearest_neighbors('T_' + str(i), vectors_k50, vocab50))
```

```
for results in results_k20:
  print(results)
```

```
for results in results_k50:
  print(results)
```
